### Q.1 Explain GET and POST methods.

* The GET method is an HTTP request method used to retrieve data from a server. It is a simple and widely used method that is used to retrieve information, such as web pages, images, videos, and other resources from the server.
When a client sends a GET request to the server, it includes a URL (Uniform Resource Locator) and any parameters that are required. The server responds with the requested data in the form of an HTTP response message.


* The POST method is an HTTP request method used to submit data to a server to create or update a resource. It is commonly used to send form data, upload files, or make requests that result in a change on the server.
When a client sends a POST request to the server, it includes the data to be submitted in the body of the request. This data can be in any format, such as JSON, XML, or plain text. Unlike the GET method, the data sent in a POST request is not visible in the URL.
The server responds to a POST request with an HTTP response message that indicates whether the request was successful or not. If the request was successful, the server may also return a response body with data about the new or updated resource.

### Q2. Why is request used in Flask?

In Flask, the request object is used to represent an HTTP request made by a client to the server. It is a global object that is available in the context of a request and can be used to access data sent by the client in the request, such as form data, headers, and cookies.

The request object allows Flask developers to create dynamic web applications that can handle a wide range of user inputs and interactions. By accessing data sent by the client, developers can create customized responses based on user inputs and interactions.

### Q3. Why is redirect() used in Flask?

In Flask, the redirect() function is used to redirect the client to a different URL or route. This function is commonly used when handling web form submissions or other types of user inputs that require a response that is not a web page.

When a client submits a web form or other input to the server, the server may need to perform some processing before returning a response. In some cases, the response may be a web page that displays the results of the processing. However, in other cases, the response may simply be a redirect to a different URL or route.

The redirect() function allows Flask developers to easily create redirects in their applications. It takes a URL or route as an argument and returns an HTTP response with a 302 status code, which indicates that the client should redirect to the specified URL or route.

### Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are used to create dynamic HTML pages that can be customized based on user input or other dynamic data. A template is a file that contains HTML code as well as placeholders for dynamic data, such as variables and control structures.

The render_template() function is used in Flask to render a Jinja2 template with dynamic data. It takes a template file name and any data that needs to be passed to the template as arguments and returns an HTTP response with the rendered HTML content.

### Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask, jsonify, request

app = Flask(__name__)

@app.route("/api", methods=['POST'])
def api():

    data = request.get_json()

    name = data['name']
    age = data['age']

    response = {'message': 'Hello, ' + name + '! You are ' + str(age) + ' years old.'}

    return jsonify(response)

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```
