### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

five plots that can be plotted using the Seaborn library and their uses:

* Scatter plot - used to visualize the relationship between two variables by plotting their values as points on a two-dimensional plane. Seaborn's scatter plot can also incorporate additional variables through the size, color, and shape of the points.
* Line plot - used to display trends in data over time or a continuous variable. Seaborn's line plot can also include confidence intervals around the line to show the uncertainty in the data.
* Histogram - used to display the distribution of a single variable. Seaborn's histogram can also show the kernel density estimate (KDE) of the data as a smooth line.
* Box plot - used to display the distribution of a variable across categories or groups. Seaborn's box plot can also show the swarm plot of the data points as individual points on top of the boxes.
* Heatmap - used to visualize the relationship between two variables as a grid of colored cells. Seaborn's heatmap can also show the values of the cells as annotations and cluster the rows and columns of the grid based on their similarity.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
fmri = sns.load_dataset('fmri')
fmri.columns
sns.lineplot(x=fmri['timepoint'], y=fmri['signal'], hue="event", style="region", data=fmri)
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
fig, axs = plt.subplots(1,2,figsize=(10,6))
sns.boxplot(ax = axs[0], x=titanic['pclass'], y=titanic['age'])
sns.boxplot(ax = axs[1], x=titanic['pclass'], y=titanic['fare'])
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
fig = plt.figure(figsize=(11,6))
sns.histplot(x = diamonds['price'], hue = diamonds['cut'], data = diamonds)
```

### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" columnof the iris dataset.

```python
iris = sns.load_dataset('iris')
fig = plt.figure(figsize=(10,6))
sns.pairplot(data =iris, hue ='species')
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Reshape the dataset for the heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Create the heatmap using Matplotlib
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu", ax=ax)

# Set the plot title and axis labels
ax.set_title("Passenger Traffic (1949-1960)", fontsize=18)
ax.set_xlabel("Year", fontsize=14)
ax.set_ylabel("Month", fontsize=14)

# Show the plot
plt.show()
```
